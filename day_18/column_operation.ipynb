# Column operations

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("C:/Users/ansh/PycharmProjects/python_learning/day_17/Sample_Data/Sample_Data.json")
```

#    Add a new column

```python
df['is_expensive']=np.where(df['price']>399,'yes','no')
df[['name','price','is_expensive']]
```

```python
# using df.insert(location,column_name,data)
df.insert(4,"price_with_tax",df['price']*1.05)
df
```

### ➕ Add Column Summary
| Method                                    | Syntax                                     | Use Case                                      | Example                                           | Notes                                       |
| ----------------------------------------- | ------------------------------------------ | --------------------------------------------- | ------------------------------------------------- | ------------------------------------------- |
| **Assign constant value**                 | `df["new_col"] = value`                    | Add a column with the same value for all rows | `df["discount"] = 10`                             | Fastest & simplest                          |
| **Assign list/array/Series**              | `df["new_col"] = [v1, v2, ...]`            | Add column with row-wise values               | `df["rating"] = [4, 5, 3, 4]`                     | Length must match number of rows            |
| **Based on another column (calculation)** | `df["new_col"] = df["col1"] * 0.1`         | Create calculated/derived columns             | `df["tax"] = df["price"] * 0.18`                  | Common in data analysis                     |
| **Using `assign()` method**               | `df = df.assign(new_col = df["col1"] + 5)` | Chain operations / method chaining            | `df = df.assign(total = df["price"] + df["tax"])` | Creates a new DataFrame (unless reassigned) |
| **Using `insert()` method**               | `df.insert(loc, "new_col", values)`        | Insert column at specific position            | `df.insert(2, "stock", [10,20,30])`               | Shifts existing columns to the right        |


# Removing Column

```python
# using df.drop(column_name,axis=1,inplace=True/False)

# if inplace true then it will remove the column permanently
# if inplace false then it will return a new dataframe with the column removed

# df.drop('price_with_tax',axis=1,inplace=True)
# df

df.drop(['id','is_expensive'],axis=1,inplace=False)
```

```python
# using df.pop(column_name) → it will remove the column permanently and return the removed column as a Series
df_copy=df.copy()
print(df_copy.pop('image'))
df_copy
```

```python
# using del keyword → it will remove the column permanently
df_copy2=df.copy()
del df_copy2['category']
df_copy2
```

## ✨ Summary Table

| Method | Removes      | Returns New DataFrame?             | Inplace? | Can remove Rows? | Returns Data?  |
| ------ | ------------ | ---------------------------------- | -------- | ---------------- | -------------- |
| `drop` | Row / Column | Yes (default), No (`inplace=True`) | Optional | ✅ Yes            | ❌ No           |
| `del`  | Column only  | ❌ No                               | ✅ Always | ❌ No             | ❌ No           |
| `pop`  | Column only  | ❌ No                               | ✅ Always | ❌ No             | ✅ Yes (Series) |

#     Rename columns

```python
# using df.rename(columns={'old_name':'new_name'},inplace=True/False)
df.rename(columns={'price_with_tax':'price_including_tax'},inplace=False)
```

```python
# using df.columns = [list of new column names]
df_copy3=df.copy()
df_copy3.columns=['ID','Product_Name','Category','Price','Price_with_Tax','Category','Image','Is_Expensive']
df_copy3
```

```python
# using df.set_axis([list of new column names],axis=1)
df.set_axis(['ID','Product_Name','Category','Price','Price_with_Tax','Category','Image','Is_Expensive'],axis=1)
```

## 📑 Rename Summary
| Method                                     | Syntax                                                                 | Use Case                         | Example                                                                       | Notes                                                |
| ------------------------------------------ | ---------------------------------------------------------------------- | -------------------------------- | ----------------------------------------------------------------------------- | ---------------------------------------------------- |
| **Single column rename (rename method)**   | `df.rename(columns={"old":"new"}, inplace=True)`                       | Rename only one column           | `df.rename(columns={"price":"cost"}, inplace=True)`                           | Flexible, can use `inplace=True`                     |
| **Multiple column rename (rename method)** | `df.rename(columns={"old1":"new1", "old2":"new2"}, inplace=True)`      | Rename multiple columns at once  | `df.rename(columns={"id":"product_id", "name":"product_name"}, inplace=True)` | Only the specified columns will change               |
| **Direct overwrite**                       | `df.columns = ["col1","col2",...]`                                     | Change all column names at once  | `df.columns = ["id","product","desc","cost","cat","img"]`                     | Old names are completely replaced; length must match |
| **String functions on columns**            | `df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")` | Clean/format column names (bulk) | `df.columns = df.columns.str.lower()`                                         | Useful for preprocessing/standardization             |
