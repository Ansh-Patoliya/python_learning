## Selecting data
       - df['column_name']
       - df[['col1','col2']]

       - df.iloc[row_index, col_index]
            - iloc → index based (rows & columns by numbers).
            - iloc[start:end] → end exclusive (normal Python slicing).
       - df.loc[row_label, col_label]
            - loc → label based (rows & columns by names).
            - loc[start:end] → end inclusive.

       - df[condition]
            - condition example:
            (1) for single condition
                    df[df['column_name'] > value]
            (2) for multiple conditions (and ( & ), or ( |))
                    df[(df['col1'] > value1) & (df['col2'] < value2)]

```python

print(df.columns)

print(df['name'].head(3)) # selecting single column

print(df['price'].head(3), df['category'].head(3)) # selecting multiple columns
#
print(df[df['price'] > 590]) # single condition
print(df[(df['price'] > 590) & (df['category']=="Electronics")]) # multiple conditions (and)
print(df[(df['price'] > 590) | (df['category']=="Electronics")]) # multiple conditions (or)
```

```python
# Part 1: Select a single row by its label (returns a Series)
print(df.loc[11])

# Part 2: Slice rows by labels 1 through 10 (end is inclusive), all columns (returns a DataFrame)
print(df.loc[1:10])

# Part 3: Select multiple non‑contiguous row labels, all columns (returns a DataFrame)
print(df.loc[[1,4,8]])

# Part 4: Select a single scalar value at row label 2 and column "name" (returns a scalar)
print(df.loc[2,'name'])

# Part 5: Slice rows by labels 2 through 10 and select only "name" and "price" columns (returns a DataFrame)
print(df.loc[2:10,['name','price']])

# Part 6: Boolean row filter on "price" and select only the "name" column (returns a Series)
print(df.loc[df['price'] > 590, 'name'])
```
