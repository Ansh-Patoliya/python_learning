# ⭐ Ranking DataFrames
    (1) rank() method
        - assigns ranks to entries in a column based on their values
        - can handle ties using different methods (average, min, max, first, dense)
        - can rank in ascending or descending order
        - syntax:
            df['rank_column'] = df['column_name'].rank(method='average', ascending=True/False)

    (2) Example usage:
        - Rank a single column:
            df['Sales_Rank'] = df['Sales'].rank(method='average', ascending=False
        - Rank with ties handled using 'min' method:
            df['Sales_Rank_Min'] = df['Sales'].rank(method='min', ascending=False)
        - Rank in ascending order:
            df['Sales_Rank_Asc'] = df['Sales'].rank(method='average',ascending=True)
        - Rank in descending order:
            df['Sales_Rank_Desc'] = df['Sales'].rank(method='average',ascending=False)

```python

'''
# Sample DataFrame
   Column Name   Non-Null Count  Dtype
---  -----------  --------------  -----
0   OrderID      100 non-null    int64
 1   Product      100 non-null    object
 2   Category     100 non-null    object
 3   Quantity     100 non-null    int64
 4   Price        100 non-null    int64
 5   Total        100 non-null    int64
 6   Region       100 non-null    object
 7   Salesperson  100 non-null    object
 8   Month        100 non-null    object
 9   Year         100 non-null    int64

practice problems(easy->difficult) according to this dataframe:
    1. Rank the DataFrame by 'Total' in descending order using the 'average' method.
    2. Rank the DataFrame by 'Region' and then by 'Salesperson' in ascending order using the 'min' method.
    3. Rank the DataFrame by 'Year' in ascending order and 'Month' in descending order using the 'dense' method.
    4. Rank the DataFrame by 'Total' with ties handled using the 'first' method.
'''
```

```python
# 1. Rank the DataFrame by 'Total' in descending order using the 'average' method.
df['Total_Rank'] = df['Total'].rank(method='average', ascending=False)
df[df['Total_Rank']<=10].sort_values(by='Total_Rank')
```

```python
# 2. Rank the DataFrame by 'Region' and then by 'Salesperson'
df['Region_Salesperson_Rank'] = df.groupby('Region')['Salesperson'].rank(method='min', ascending=True)
df.head()
```

```python
# 3. Rank the DataFrame by 'Year' in ascending order and 'Month'
df['Year_Month_Rank'] = df.groupby('Year')['Month'].rank(method='dense', ascending=False)
df.head()
```

```python
# 4. Rank the DataFrame by 'Total' with ties handled using the 'first
df['Total_Rank_First'] = df['Total'].rank(method='first', ascending=False)
df.head()
```
