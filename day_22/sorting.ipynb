# ⭐ Sorting DataFrames
    (1) sort_values() method
        - by default, sorts in ascending order
        - can sort by multiple columns
        - can specify ascending or descending order for each column
        - can handle missing values (NaN) by placing them at the beginning or end of the
        sorted DataFrame
        - syntax:
          df.sort_values(by='column_name', ascending=True/False, na_position='first'/'last')

    (2) sort_index() method
        - sorts the DataFrame based on its index
        - can sort in ascending or descending order
        - syntax:
            df.sort_index(ascending=True/False)

    (3) inplace parameter
        - if set to True, modifies the original DataFrame
        - if set to False (default), returns a new sorted DataFrame without modifying the
        original

    (4) Example usage:
        - Sort by a single column:
            df_sorted = df.sort_values(by='Sales', ascending=False)
        - Sort by multiple columns:
            df_sorted = df.sort_values(by=['Region', 'Sales'], ascending=[True, False])
        - Sort by index:
            df_sorted = df.sort_index(ascending=True)
        - Sort with missing values at the end:
            df_sorted = df.sort_values(by='Sales', na_position='last')

```python
'''
# Sample DataFrame
   Column Name   Non-Null Count  Dtype
---  -----------  --------------  -----
0   OrderID      100 non-null    int64
 1   Product      100 non-null    object
 2   Category     100 non-null    object
 3   Quantity     100 non-null    int64
 4   Price        100 non-null    int64
 5   Total        100 non-null    int64
 6   Region       100 non-null    object
 7   Salesperson  100 non-null    object
 8   Month        100 non-null    object
 9   Year         100 non-null    int64

practice problems(easy->difficult) according to this dataframe:
    1. Sort the DataFrame by 'Total' in descending order.
    2. Sort the DataFrame by 'Region' and then by 'Salesperson' in ascending order.
    3. Sort the DataFrame by 'Year' in ascending order and 'Month' in descending order.
    4. Sort the DataFrame by 'Total' with missing values (NaN) placed at the end.

'''

# 4.Sort the DataFrame by 'Total' with missing values (NaN) placed at the end.
df.sort_values(by='Total',ascending=True,inplace=False)
```

```python
# 3. Sort the DataFrame by 'Year' in ascending order and 'Month' in descending order.
df.sort_values(by=['Year','Month'] , ascending=[True,False],inplace=False)
```

```python
# 2. Sort the DataFrame by 'Region' and then by 'Salesperson' in ascending order.
df.sort_values(by=['Region','Salesperson'],ascending=[True,True],inplace=False)
```

```python
# 1. Sort the DataFrame by 'Total' in descending order.
df.sort_values(by='Total',ascending=False,inplace=False).reset_index()
```
