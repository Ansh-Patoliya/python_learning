```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 8, 15, 3, 12]

plt.plot(x, y)
plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
company_A_price = [100, 105, 110, 108, 115]
company_B_price = [90, 92, 98, 105, 102]

plt.plot(days, company_A_price, label='Company A Stock')
plt.plot(days, company_B_price, label='Company B Stock')

plt.title("Stock Prices Over 5 Days")
plt.xlabel("Day")
plt.ylabel("Price (in USD)")

plt.legend()

plt.savefig('stock_prices_plot.png', dpi=300)

plt.show()
```
