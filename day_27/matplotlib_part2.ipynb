```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('matches.csv',encoding='latin1')
total_match=df.groupby('season')['id'].count()
# print(total_match)

plt.figure(figsize=(12, 7))
plt.plot(total_match.index,total_match.values,
         marker='o',
         linestyle='--',
         color='#1E90FF',
         linewidth=3
         )
plt.xlabel('Seasons',fontsize='15')
plt.ylabel("Total Matches",fontsize='15')
plt.title('Matches per season',fontsize='18')
plt.xticks(total_match.index,rotation=45)
plt.grid(True,linestyle='--',alpha=0.7)
plt.show()
```

```python
import seaborn as sns

sns.set_style('darkgrid')
plt.figure(figsize=(12,7))
sns.lineplot(x=total_match.index,
             y=total_match.values,
             marker='o',
             linewidth=3
             ,color='orange')
plt.xlabel('Seasons',fontsize='15')
plt.ylabel("Total Matches",fontsize='15')
plt.title('Matches per season',fontsize='18')
plt.xticks(total_match.index,rotation=45)
plt.savefig('lineplot_season.png',dpi=300)
plt.show()
```

```python
winners=df.groupby('winner')['id'].count().sort_values(ascending=False)

sns.set_style('darkgrid')
plt.figure(figsize=(12,7))
sns.lineplot(
    x=winners.index,
    y=winners.values
)
plt.xticks(winners.index,rotation=90)
plt.show()
```

```python


# remove 2021 to 2024 but season is str
df=df[~df['season'].isin(['2021','2022','2023','2024'])]


mom=df.groupby('player_of_match')['id'].count()
mom=mom.sort_values(ascending=False).head()

sns.set_style('darkgrid')
plt.figure(figsize=(12,7))
sns.lineplot(
    x=mom.index,
    y=mom.values
)
plt.xlabel('Player Name')
plt.ylabel('Man of the match')
plt.title('Most mom award')
# plt.xticks(mom.index,45)

plt.show()
```

```python
mm=df.groupby('venue')['id'].count().sort_values(ascending=False).head(10)
sns.set_style('darkgrid')
plt.figure(figsize=(12,7))
sns.lineplot(
    x=mm.index,
    y=mm.values,
    marker='o',
    linewidth=2,
    color='green'
)
plt.xlabel('Venue')
plt.ylabel('Matches Played')
plt.title('Top 10 Venue with most matches')
plt.xticks(mm.index,rotation=45)
plt.show()
```
